```python
import geopandas as gpd

year = 2020
v = 1
gdf = gpd.read_parquet(f'../data/results/disturbances_{year}_v{v}.parquet')
```

```python
groups = gdf.groupby('index_ref')
from itertools import islice 

g = []
for name, group in islice(groups,1000):
    g.append(group)
```

```python
[i for i,r in enumerate(g) if len(r) > 10 ]
```

```python
data = g[912]

try:
    data['start_date'] = data['start_date'].dt.strftime('%Y-%m-%d')
except :
    pass 

try:
    data['end_date'] = data['end_date'].dt.strftime('%Y-%m-%d')
except : 
    pass

data.explore(column = 'dataset', 
             tooltip = ['year', 'start_date', 'end_date', 'essence', 'class', 'notes'], 
             tiles = 'Esri.WorldImagery',
             marker_kwds=dict(radius=5, fill=True))
```
