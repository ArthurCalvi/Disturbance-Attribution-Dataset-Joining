```python
import geopandas as gpd 

polygons = []
for year in range(2017,2021):
    polygons.append(gpd.read_parquet(f'/Users/arthurcalvi/Repo/Disturbance-Attribution-Dataset-Joining/data/results/cluster_{year}/clusters_filtered_{year}_g5_v0.3_polygons.parquet'))

import pandas as pd
polygons = gpd.GeoDataFrame(pd.concat(polygons, ignore_index=True))
```

```python
classes = polygons['Class'].unique()

#sample 10 polygons from each class
sample = polygons.groupby('Class').apply(lambda x: x.sample(10))
```
