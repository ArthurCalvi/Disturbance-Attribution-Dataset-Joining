## Gating functions

```python
import numpy as np
import matplotlib.pyplot as plt

#set font to Times New Roman
plt.rcParams["font.family"] = "Times New Roman"

# Define a Gaussian function
def gaussian(x, mean, std):
    return np.exp(-0.5 * ((x - mean) / std)**2)

# Define a step function
def step(x, start, end):
    return np.where((x >= start) & (x <= end), 1, 0)

# Define a sigmoid function
def sigmoid(x, x0, k):
    return 1 / (1 + np.exp(-k * (x - x0)))

# Define an exponential decay function
def exponential_decay(x, x0, k):
    return np.exp(-k * (x - x0))

# Define a constant function
def constant(x, value):
    return np.full_like(x, value)

# Define gating functions for each disturbance
def fire_gating(x_spatial, x_temporal):
    # Spatial: Gaussian, Temporal: Step
    spatial = gaussian(x_spatial, mean=0, std=250)  # Example parameters
    temporal = step(x_temporal, start=0, end=365)     # Example parameters
    return spatial, temporal

def storm_wind_gating(x_spatial, x_temporal):
    # Spatial: Gaussian (elongated), Temporal: Step
    spatial = gaussian(x_spatial, mean=0, std=500)  # Elongated Gaussian
    temporal = step(x_temporal, start=0, end=365)      # Shorter duration
    return spatial, temporal

def biotic_dieback_gating(x_spatial, x_temporal):
    # Spatial: Sigmoid, Temporal: Sigmoid or Exponential Decay
    spatial = gaussian(x_spatial, mean=0, std=1000)     # Gradual onset
    temporal = exponential_decay(x_temporal, x0=0, k=0.0025)  # Longer duration
    return spatial, temporal

def drought_dieback_gating(x_spatial, x_temporal):
    # Spatial: Constant, Temporal: Sigmoid or Exponential Decay
    spatial = gaussian(x_spatial, mean=0, std=2500)          # Uniform impact
    temporal = gaussian(x_spatial, mean=0, std=1.5*365) # Long duration
    return spatial, temporal

def mortality_gating(x_spatial, x_temporal):
    # Spatial: Gaussian or Sigmoid, Temporal: Exponential Decay
    spatial = gaussian(x_spatial, mean=0, std=100)  # Depends on preceding dieback
    temporal = exponential_decay(x_temporal, x0=0, k=0.005)  # Gradual decrease
    return spatial, temporal

def anthropogenic_gating(x_spatial, x_temporal):
    # Spatial: Step or Sharp Gaussian, Temporal: Step or Sigmoid/Exponential Decay
    spatial = step(x_spatial, start=0, end=250)     # Sharp boundaries
    temporal = step(x_temporal, start=0, end=365)       # Combination of sudden and gradual
    return spatial, temporal

# Example usage
x_spatial = np.linspace(0, 7000, 1000)
x_temporal = np.linspace(0, 365*5, 1000) # Example range

dgatings = {
    'Fire': fire_gating,
    'Storm': storm_wind_gating,
    'Biotic-Dieback': biotic_dieback_gating,
    'Drought-Dieback': drought_dieback_gating,
    'Mortality': mortality_gating,
    'Anthropogenic': anthropogenic_gating
}

fig, axes = plt.subplots(2,6, figsize=(10, 4), sharey=True)
for i, (t, g) in enumerate(dgatings.items()):
    ax = axes[:,i]
    ax[0].set_title(t)

    spatial, temporal = g(x_spatial, x_temporal)
    ax[0].plot(x_spatial, spatial, label="Spatial", color='blue', linewidth=2)
    ax[0].set_xlabel("Spatial Distance [m]", fontsize=7)

    ax[1].plot(x_temporal, temporal, label="Temporal", color='blue', linewidth=2)
    ax[1].set_xlabel("Temporal Distance [d]", fontsize=7)

    if i == 0 :
        ax[0].set_ylabel("Spatial Weight [1]", fontsize=7)
        ax[1].set_ylabel("Temporal Weight [1]", fontsize=7)


    for a in ax:
        a.grid()
        a.spines['top'].set_visible(False)
        a.spines['right'].set_visible(False)

plt.tight_layout()

```
